```python
import pandas as pd
import numpy as np
```

```python
nyc = pd.read_csv('../data/newyork_housing.csv')
```

```python
nyc = nyc.drop(nyc.filter(like='photos',axis=1), axis=1)
```

```python
#Dropping all reso columns if they have more than 50% missinig data
for i in nyc.columns:
    if 'reso' in i:
        if nyc[i].isnull().sum() > 40000:
            nyc = nyc.drop(i,axis=1)
            print(i)
        else:
            continue
    else:
        continue
    
```

```python
missing_data_cols  = []
for i in nyc.columns:
    if nyc[i].isnull().sum() >= 40000:
        missing_data_cols.append(i)
    else:
        continue
    
        
```
