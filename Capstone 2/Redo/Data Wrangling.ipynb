## Dropping Columns

```python
non_ny_house = df[df['address/state'] != 'NY'].index
df = df.drop(non_ny_house, axis=0).reset_index(drop=True)
```

```python
drop_columns = missing[missing['%'] > 50].index
```

```python
pd.set_option('display.max_rows', 20)
```

```python
df = df.drop(drop_columns, axis=1)
```

```python
reso_drop_list = [] 
for i in df.columns:
    if 'reso' in i:
        reso_drop_list.append(i)
        
len(reso_drop_list)
```

```python
reso_list = ['resoFactsStats/atAGlanceFacts/0/factValue', 'resoFactsStats/atAGlanceFacts/1/factValue', 
             'resoFactsStats/atAGlanceFacts/4/factValue',
            'resoFactsStats/atAGlanceFacts/7/factLabel','resoFactsStats/atAGlanceFacts/7/factValue', 'resoFactsStats/basement', 'resoFactsStats/bathrooms', 'resoFactsStats/bathroomsFull', 
            'resoFactsStats/bathroomsHalf', 'resoFactsStats/bedrooms','resoFactsStats/homeType', 'resoFactsStats/livingArea', 'resoFactsStats/lotSize', 'resoFactsStats/parkingFeatures/0',
            'resoFactsStats/parkingFeatures/1', 'resoFactsStats/stories', 'resoFactsStats/taxAnnualAmount', 'resoFactsStats/taxAssessedValue',
             'resoFactsStats/yearBuilt']
```

```python
reso_drop = [x for x in reso_drop_list if x not in reso_list]
#Making a new dataframe of only reso_list so I can drop all reso columns and re-merge with the reso columns
#we want
#Drop all columns with reso
df = df.drop(reso_drop,axis=1)
```

```python
drop_rows_wo_price = df[df.price.isnull()].index
```

```python
df = df.drop(drop_rows_wo_price, axis=0).reset_index(drop=True)
```

```python
df["resoFactsStats/bedrooms"].fillna(df["bedrooms"], inplace=True)
df["bathrooms"].fillna(df["resoFactsStats/bathrooms"], inplace=True)
df['resoFactsStats/atAGlanceFacts/1/factValue'].fillna(df['resoFactsStats/yearBuilt'],inplace=True)
df['livingArea'].fillna(df['resoFactsStats/livingArea'], inplace=True)
```

```python
drop_col = ['bathrooms', 'bedrooms', 'zpid', 'url', 'currency', 'address/state'
             ,'resoFactsStats/parkingFeatures/0', 'homeStatus', 'description'
             , 'resoFactsStats/bathrooms', 'bedrooms', 'resoFactsStats/homeType'
             ,'resoFactsStats/livingArea', 'resoFactsStats/yearBuilt', 'yearBuilt']
```

```python
for i in df.columns[df.columns.str.contains('photo')].values:
    drop_col.append(i)

```

```python
priceHist_cols = []
for i in df.columns[df.columns.str.contains('priceHistory')]:
    priceHist_cols.append(i)
```

```python
column_rename = {'address/city':'City', 'address/zipcode':'ZipCode', 
                 'resoFactsStats/atAGlanceFacts/0/factValue': 'HomeType', 
                 'resoFactsStats/atAGlanceFacts/1/factValue':'YearBuilt', 
                 'resoFactsStats/atAGlanceFacts/4/factValue': 'ParkingType', 'priceHistory/0/priceChangeRate':'priceChangeRate',
                'resoFactsStats/basement':'Basement', 'bathrooms': 'Total_Bathrooms',
                 'resoFactsStats/bathroomsFull':'Full_Bathrooms','resoFactsStats/bathroomsHalf':'Half_Bathrooms',
                 'resoFactsStats/bedrooms':'Bedrooms', 
                 'resoFactsStats/lotSize':'Lot_Size', 'resoFactsStats/stories':'Story_Building',
                 'resoFactsStats/taxAnnualAmount':'Annual_Tax', 'resoFactsStats/taxAssessedValue':'Tax_Assessed_Value',
                'schools/0/distance':'school_1_distance', 'schools/0/grades':'school_1_grades', 'schools/0/level':'school_1_level',
                'schools/0/rating': 'school_1_rating', 'schools/0/size': 'school_1_size', 'schools/0/studentsPerTeacher':'school_1_s/t_ratio',
       'schools/0/type':'school_1_type', 'schools/1/distance':'school_2_distance', 'schools/1/level':'school_2_level',
       'schools/1/rating': 'school_2_rating', 'schools/1/size': 'school_2_size', 'schools/1/studentsPerTeacher': 'school_2_s/t_ratio',
       'schools/1/type':'school_2_type'}
df.rename(columns = column_rename, inplace = True)
```

```python
def to_sqft(a):
    return float(a*43560)

df1 = df[['Lot_Size']][df['Lot_Size'].str.contains(r'Acres', na=False)]
df1['Lot_Size'] = df1['Lot_Size'].replace(to_replace ='Acres', value = '', regex = True)
#Get rid of sqft as we denouce it was in SqFT in the column name
df1['Lot_Size'] = df1['Lot_Size'].replace(to_replace =' ', value = '', regex = True)
df1['Lot_Size'] = df1['Lot_Size'].replace(to_replace =',', value = '', regex = True)

df1['Lot_Size'] = df1['Lot_Size'].astype(float)

df['Lot_Size'][df['Lot_Size'].str.contains(r'Acres', na=False)] = df1['Lot_Size'].apply(to_sqft)
```

```python
df['Lot_Size'] = df['Lot_Size'].replace(to_replace ='sqft', value = '', regex = True)
#Get rid of sqft as we denouce it was in SqFT in the column name
df['Lot_Size'] = df['Lot_Size'].replace(to_replace =' ', value = '', regex = True)
df['Lot_Size'] = df['Lot_Size'].replace(to_replace =',', value = '', regex = True)

df['Lot_Size'] = df['Lot_Size'].astype(float)
```

```python
df['livingArea'] = df['livingArea'].replace(to_replace ='sqft', value = '', regex = True)
#Get rid of sqft as we denouce it was in SqFT in the column name
df['livingArea'] = df['livingArea'].replace(to_replace =' ', value = '', regex = True)
df['livingArea'] = df['livingArea'].replace(to_replace =',', value = '', regex = True)

df['livingArea'] = df['livingArea'].astype(float)
```

```python
duplicates = df.duplicated(subset = ['address/streetAddress', 'price', 'latitude', 'longitude', 'YearBuilt'], keep=False)
```

```python
df.drop_duplicates(subset=['address/streetAddress', 'price', 'latitude', 'longitude', 'YearBuilt'], inplace=True)
df = df.reset_index(drop=True)
```

```python
vacant_lands = df[df['HomeType'] == 'Vacant Land'].index
df = df.drop(vacant_lands, axis=0)
df = df.reset_index(drop=True)
```

```python
df = df.drop(df[df['price'] < 100000].index, axis=0)
df = df.drop(df[df.price == 8250000].index, axis=0).reset_index(drop=True)
df = df.drop(df[df.price > 10000000].index, axis=0).reset_index(drop=True)
```
