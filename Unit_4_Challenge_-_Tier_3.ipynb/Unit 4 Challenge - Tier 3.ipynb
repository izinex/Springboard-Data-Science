**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(d):
    y1998 = float(d['Average_price'][d['Year']==1998])
    y2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [y2018/y1998]
    return ratio
```

```python
#make an empty dict to hold all of our ratios of the boroughs between 1998 and 2018

ratio_dif = {}
price_1998 = {}
price_2018 = {}

each_borough = df['London_Borough'].unique()
```

```python
for p in each_borough:
    borough = df[(df['London_Borough'] == p) & ((df['Year'] == 1998) | (df['Year'] == 2018))]
    price_1998[p] = [df.at[df[(df['London_Borough'] == p) & (df['Year'] == 1998)].index[0],'Average_price']]
    price_2018[p] = [df.at[df[(df['London_Borough'] == p) & (df['Year'] == 2018)].index[0],'Average_price']]
    
    

```

```python
# We will iterate using a for loop to put all the ratios inside ratio_dif dict

for x in each_borough:
    borough = df[df['London_Borough'] == x]
    ratio_dif[x] = create_price_ratio(borough)
    
print(ratio_dif)
```

```python
#Create a dataframe for our price in 1998 and 2018
df_price_1998 = pd.DataFrame(price_1998)
df_price_2018 = pd.DataFrame(price_2018)
```

```python
df_price_1998_T = df_price_1998.T
df_price_1998 = df_price_1998_T.reset_index()
df_price_1998.head()
```

```python
df_price_2018_T = df_price_2018.T
df_price_2018 = df_price_2018_T.reset_index()
df_price_2018.head()
```

```python
df_price_1998.rename(columns={'index':'Borough', 0:'Y_1998'}, inplace=True)
df_price_2018.rename(columns={'index':'Borough', 0:'Y_2018'}, inplace=True)
df_price_1998.head()
```

```python
df_price_1998['Y_2018'] = df_price_2018['Y_2018']
final_df_price_diff = df_price_1998
final_df_price_diff
```

```python
# Create a dataframe with our new ratios so we can make plots using that dataframe
df_ratio = pd.DataFrame(ratio_dif)
```

```python
df_ratio_t = df_ratio.T
df_ratios = df_ratio_t.reset_index()
df_ratios.head()
```

```python
df_ratios.rename(columns={'index':'Borough', 0:'2018'}, inplace=True)
df_ratios.head()
```

```python
# Add the difference in ratio column to the Final Data frame
final_df_price_diff['ratio'] = df_ratios['2018']
```

```python
final_df_price_diff.head()
```

```python
#Now we can see which borough had the highest average price change over the two decades between 1998-2018
print(final_df_price_diff.sort_values('ratio', ascending=False))
top5 = final_df_price_diff.sort_values('ratio', ascending=False).head(5)
# we can see Hackney had the highest price difference since 1998 
```

```python
top5price = top5[['Y_1998','Y_2018', 'ratio']].plot(kind='bar')
plt.yscale('log')
top5price.set_xticklabels(top5.Borough)
```
