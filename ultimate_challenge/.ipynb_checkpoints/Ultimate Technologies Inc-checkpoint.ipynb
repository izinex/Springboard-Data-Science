## Part 2 - Experiments and metric design

### 1) What would you choose as the key measure of success of this experiment in encouraging driver partners to serve both cities, and why would you choose this metric?

A key measure of sucess can be seen by the activity level throughout both cities especially during their off-peak hours, since they are complementary where Gotham is active at night and Metropolis is active at day, if both cities see a spike on their off-peak hours, for Gotham that would be during the day, and Metropolis during the night, this can be used as a key measure for success. While one can see the reimbursement data to see if drivers are taking the bridge between the two cities, I don't believe that data shows true correlation of availibility to both cities, as drivers could easily utilize the bridges to drop off one customer for additional fare but return to their own city, thus I believe the activity level is a better indicator.

### 2) Describe a practical experiment you would design to compare the effectiveness of the proposed change in relation to the key measure of success. Please provide details on:

A practical experiment that could suffice is implementing the proposed change on certain weeks, and seeing if our key measure of success changes from the different weeks where toll fares are reimbursed versus weeks where it isn't. If the data shows there are more activity level during off-peak hours at both cities during weeks where Ultima technology offers toll reimbursement, versus week it isn't. 

### a) how you will implement the experiment

As stated above, we can simply track the activity level of both cities through the day and compare their off-peak hours, to see if there was any change, and if it was significant enough. To do so we would alternate or choose random weeks where tolls would be reimbursed and the other weeks keep everything standard.

### b) what statistical test(s) you will conduct to verify the significance of the observation


We can use Hypothesis testing to assess our data and what it shows, our Null Hypothesis or H0 would be, there is no effect of on activity levels of the two cities during off-peak hours given our proposed change. Our Alternate Hypothesis or HA is, there is change of activity levels of the two cities during off-peak hours given our proposed change. 

### c) how you would interpret the results and provide recommendations to the city operations team along with any caveats.

Depending on our results I would assess the significance between the increase or lack of increase in activity levels during off-peak hours between the two cities, and recommend to city operations to continue the toll reimbursement if there was a significant increase in activity levels, or if there is no significance increase, revise or try a different approach like increasing car fare or giving city benefits during their off-peak hours.

## Modeling

Our objective is to create an model which can predict if a user will stay active thus it is a classification problem, there are many ways I can take this approach, I can attempt a logistic regression, decission tree, gradient boost, or even nearest neighbors. I will attempt many models and utilize the model that gives be the best predictions. In doing so I will extract the most important features that leads to active users. 

In order to get predictions I will be splitting our dataset into training and testing split. 

```python
from sklearn import neighbors, datasets, preprocessing
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import r2_score as r2

from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV, learning_curve
```

```python
X = ultimate.drop(['active', 'signup_date', 'last_trip_date', 'days_used'], axis=1)

y = ultimate["active"]
                  

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=7)
```

### Logistic Regression

#### Linear and Logistic Regression does not work with datetime data, we must either transform our dates to numerical data or drop it. While I do not believe dates effect whether or not a user is 'active' we must first check before we can rule it out

```python
from sklearn.linear_model import LogisticRegression

logit_reg = LogisticRegression(max_iter=1000, solver='liblinear', penalty='l1')
```

```python
logit_reg.fit(X_train, y_train)
```

```python
y_pred = logit_reg.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(logit_reg.coef_)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

We can see that Phone platform was the the most impactful, with being an ultimate black user being the next most impactful, after those two the rest are not as impactful but the number of days a user used ultimate had a high impact as well, as well as the city.

Hyper-parameter Tuning for XGBoost

```python
param_grid = {
        'max_depth': [6, 10, 15, 20],
        'learning_rate': [0.001, 0.01, 0.1, 0.2, 0,3],
        'subsample': [0.5, 0.6, 0.7, 0.8, 0.9, 1.0],
        'colsample_bytree': [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0],
        'colsample_bylevel': [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0],
        'min_child_weight': [0.5, 1.0, 3.0, 5.0, 7.0, 10.0],
        'gamma': [0, 0.25, 0.5, 1.0],
        'reg_lambda': [0.1, 1.0, 5.0, 10.0, 50.0, 100.0],
}
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Init classifier
xgb_cl = xgb.XGBClassifier(objective="binary:logistic")

# Init Grid Search
rn_cv = RandomizedSearchCV(xgb_cl, param_grid, n_jobs=-1, cv=3, scoring="accuracy")

# Fit
_ = rn_cv.fit(X_train, y_train)
```

```python
rn_cv.best_score_
```

```python
rn_cv.best_params_
```

```python
final_xg = xgb.XGBClassifier(
    **rn_cv.best_params_,
    objective="binary:logistic",
)
```

```python
final_xg.fit(X_train, y_train)

final_preds = final_xg.predict(X_test)
```

```python
accuracy_score(y_test, final_preds)
```

```python
print(confusion_matrix(y_test, final_preds))
```

```python
print(classification_report(y_test, final_preds))
```

```python
final_xg.feature_importances_
```
